```python
def plot_var(usuari, any, mes, dia, colors, vars, text):
  df_plot = df[(df['Usuari'] == usuari) & (df['Any'] == any) & (df['Mes'] == mes) & (df['Dia'] == dia)]
  fig, ax = plt.subplots()
  temps = df_plot['P1 Temps Sistema']
  #temps = df_plot['P2 Temps Sistema']
  for i, columna in enumerate(df_plot.columns[vars]):
      #print(columna)
      ax.plot(temps, df_plot[columna], label=columna, color=colors[i])
  #ax.legend()

  idx = np.arange(0,len(temps),2000)
  selected_dates = temps.iloc[idx]

  # Generar el gráfico de tiempo vs. sensor1
  plt.xlabel('Temps')
  plt.ylabel(text)
  plt.title(str("Gràfic del temps vs. "+text+" per l'usuari " + str(usuari)))
  plt.xticks(selected_dates, rotation=45)
  #plt.ylim(0,100)
  #plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
  plt.legend()
  plt.show()
```

```python
colors = ['blue', 'green', 'red', 'cyan', 'orange', 'pink', 'purple', 'turquoise', 'gray', 'brown', 'black', 'silver', 'gold', 'yellow']
```
